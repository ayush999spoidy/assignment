10.Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a

```python
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)
```

```python
size = 30
feature1 = np.random.normal(0, 1, size)
feature2 = 2 * feature1 + np.random.normal(0, 1, size)
data = pd.DataFrame({
    'Feature 1': feature1,
    'Feature 2': feature2})
```

6.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")
output_notebook()
show(p)
```

```python
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
from bokeh.models import ColorBar
```

```python
np.random.seed(20)
data_heatmap = np.random.rand(10, 10)
```

```python
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
```

```python
xx_flat = xx.flatten()
yy_flat = yy.flatten()
data_flat = data_heatmap.flatten()
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1), width=400, height=400)
```

```python
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(data_flat), high=max(data_flat))
```

```python
color_bar = ColorBar(color_mapper=mapper['transform'], location=(0, 0))
p.add_layout(color_bar, 'right')
output_notebook()
show(p)
```
