12.Write a Python program to train a Decision Tree Classifier, apply Cost Complexity Pruning (CCP), and
visualize its effect on accuracy.

```python
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X, y = iris.data, iris.target
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf = DecisionTreeClassifier(random_state=0)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.fit(X_train, y_train)
```

```python
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
```

```python
accuracies = []
for ccp_alpha in ccp_alphas:
    clf_pruned = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    clf_pruned.fit(X_train, y_train)
    y_pred = clf_pruned.predict(X_test)
    accuracies.append(accuracy_score(y_test, y_pred))
```

```python
plt.plot(ccp_alphas, accuracies, marker='o', drawstyle="steps-post")
plt.xlabel("effective alpha")
plt.ylabel("accuracy")
plt.title("Accuracy vs effective alpha for training and testing sets")
plt.show()
```

15.Write a Python program to train a Decision Tree Classifier and use GridSearchCV to find the optimal values
for max_depth and min_samples_split.

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
iris = load_iris()
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)
```

```python
param_grid = {
    'max_depth': [2, 3, 4, 5, 6],
    'min_samples_split': [2, 5, 10, 15, 20]
}
```

```python
dtc = DecisionTreeClassifier()
```

```python
grid_search = GridSearchCV(estimator=dtc, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```

```python
best_dtc = DecisionTreeClassifier(**grid_search.best_params_)
best_dtc.fit(X_train, y_train)
```

```python
y_pred = best_dtc.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy on test set:", accuracy)
```
